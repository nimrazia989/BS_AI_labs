# Task 02 – Simple Calculator

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operation = input("Choose operation (+, -, *, /, **, %): ")

if operation not in ["+", "-", "*", "/", "**", "%"]:
    print("Invalid operator!")

elif operation == "/" and num2 == 0:
    print("Cannot divide by zero!")

else:
    if operation == "+":
        result = num1 + num2
        print(f"{num1} + {num2} = {result}")

    elif operation == "-":
        result = num1 - num2
        print(f"{num1} - {num2} = {result}")

    elif operation == "*":
        result = num1 * num2
        print(f"{num1} * {num2} = {result}")

    elif operation == "/":
        result = num1 / num2
        print(f"{num1} / {num2} = {result:.2f}")

    elif operation == "**":
        result = num1 ** num2
        print(f"{num1} ** {num2} = {result:.2f}")

    elif operation == "%":
        result = num1 % num2
        print(f"{num1} % {num2} = {result}")

# Task 03 Even/Odd & Multiple Checker


```python
start = int(input("Enter starting number: "))
end = int(input("Enter ending number: "))

num = start

while num <= end:

    if num % 7 == 0:
        num += 1
        continue

    if num % 2 == 0 and num % 5 == 0:
        print(f"{num} → Even & Multiple of 5!!")

    elif num % 2 == 0:
        print(f"{num} → Even")

    elif num % 5 == 0:
        print(f"{num} → Multiple of 5!")

    else:
        print(num)

    num += 1
```

# TASK 04 Password Strength Checker

```python
password = input("Enter a password: ")

if len(password) < 6:
    print("Too short!")

else:
    has_digit = any(char.isdigit() for char in password)
    has_upper = any(char.isupper() for char in password)

    if not has_digit:
        print("Add at least one number")

    elif not has_upper:
        print("Add at least one capital letter")

    else:
        print("Strong password – good choice!")
```
