```python
import random

secret = random.randint(1, 50)
max_attempts = 7
attempts_taken = 0

print("--- Welcome to the Neural Link Number Guesser ---")
print(f"I've picked a number between 1 and 50. You have {max_attempts} tries.")
print("-" * 45)

while attempts_taken < max_attempts:
    remaining = max_attempts - attempts_taken
    print(f"\nAttempts remaining: {remaining}")
    
    user_input = input("Enter your guess (1-50): ")
    
    if not user_input.isdigit():
        print("Invalid input! Please enter a whole number.")
        continue
    
    guess = int(user_input)
    
    if guess < 1 or guess > 50:
        print("Out of range! Please stay between 1 and 50.")
        continue
    
    attempts_taken += 1
    
    if guess == secret:
        print(f"\nCorrect! You win in {attempts_taken} attempts!")
        print("AI training level: Beginner -> Intermediate")
        break
    elif guess > secret:
        print(f"Too high! {guess} is above the target.")
    else:
        print(f"Too low! {guess} is below the target.")

else:
    print("\n" + "=" * 45)
    print(f"Game Over! You've exhausted your attempts.")
    print(f"The secret number was: {secret}")
    print("=" * 45)
```
