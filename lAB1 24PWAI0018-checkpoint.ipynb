# Data Type Program

```python
# Age after 10 years
age = int(input("Enter your age: "))
print("After 10 years, you will be:", age + 10)

# Average marks
marks1 = float(input("Enter marks 1: "))
marks2 = float(input("Enter marks 2: "))
marks3 = float(input("Enter marks 3: "))

average = (marks1 + marks2 + marks3) / 3
print("Average marks:", average)
```

# Conditions & Loops

```python
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("Even Number")
else:
    print("Odd Number")
```

# Grade System

```python
marks = int(input("Enter your marks: "))

if marks >= 80:
    print("Grade A")
elif marks >= 60:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")
```

# Multiplication Table

```python
num = int(input("Enter a number: "))

for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```

# Sum of First 10 Numbers

```python
total = 0

for i in range(1, 11):
    total += i

print("Sum of first 10 numbers:", total)
```

# Function (Average)

```python
def average(a, b, c):
    return (a + b + c) / 3

result = average(10, 20, 30)
print("Average is:", result)
```
